```python
# Memulai Spark session
spark = SparkSession.builder.appName("BigDataPractice").getOrCreate()

# Membuat DataFrame sederhana
data = [("Ali", 34), ("Budi", 23), ("Citra", 29), ("Dina", 45)]
columns = ["Nama", "Usia"]
df = spark.createDataFrame(data, columns)

# Menampilkan DataFrame
df.show()

# Filtering data
df_filtered = df.filter(df['Usia'] > 30)
df_filtered.show()

# Menghitung rata-rata usia
from pyspark.sql.functions import avg
df.groupBy().agg(avg("Usia")).show()

# Mengurutkan data berdasarkan usia
df_sorted = df.orderBy("Usia", ascending=False)
df_sorted.show()
```

Tugas 2: Lakukan filter, penghitungan rata-rata, dan pengurutan data menggunakan PySpark.

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import avg, max, min, count, col

# Memulai Spark session
spark = SparkSession.builder.appName("BigDataPractice").getOrCreate()

# Dataset dengan Pekerjaan dan Hobi yang lebih beragam
data_tugas = [
    ("Ali", 34, "Nelayan", "Memancing", "Laki-laki"),
    ("Budi", 23, "Buruh", "Bersepeda", "Laki-laki"),
    ("Citra", 29, "Guru", "Menonton Drakor", "Perempuan"),
    ("Dina", 45, "Akuntan publik", "Joging", "Perempuan"),
    ("Eka", 31, "Koki", "Memasak", "Perempuan"),
    ("Fajar", 27, "Buruh", "Memancing", "Laki-laki"),
    ("Galih", 38, "Insinyur", "Mendaki Gunung", "Laki-laki"),
    ("Hani", 22, "Penjahit", "Bersepeda", "Perempuan"),
    ("Indra", 41, "Nelayan", "Membaca Buku", "Laki-laki"),
    ("Joko", 55, "Buruh", "Joging", "Laki-laki"),
    ("Kartika", 25, "Desainer Grafis", "Fotografi", "Perempuan"),
    ("Lukman", 30, "Pelukis", "Melukis", "Laki-laki")
]

columns = ["Nama", "Usia", "Pekerjaan", "Hobi", "Gender"]
df_tugas = spark.createDataFrame(data_tugas, columns)

print("DataFrame Awal:")
df_tugas.show(n=len(data_tugas))

#Filtering Data
print("DataFrame yang difilter (Usia 25-40):")
df_filtered1 = df_tugas.filter((df_tugas['Usia'] > 25) & (df_tugas['Usia'] < 40))
df_filtered1 = df_filtered1.orderBy("Usia", ascending=True)
df_filtered1.show()

print("DataFrame yang difilter (40<Usia<30)")
df_filtered2 = df_tugas.filter((df_tugas['Usia'] > 40) | (df_tugas['Usia'] < 30))
df_filtered2 = df_filtered2.orderBy("Usia", ascending=True)
df_filtered2.show()

print("DataFrame yang difilter (Gender Laki-laki):")
df_filtered3 = df_tugas.filter(df_tugas['Gender'] == "Laki-laki")
df_filtered3.show()

#Agregasi Data
print("Hasil Agregasi:")
df_tugas.agg(
    avg("Usia").alias("Rata-rata Usia"),
    max("Usia").alias("Usia Tertua"),
    min("Usia").alias("Usia Termuda")
).show()

#Sorting Data
print("DataFrame yang diurutkan (dari yang tertua):")
df_sorted_tugas = df_tugas.orderBy("Usia", ascending=False)
df_sorted_tugas.show()
```

Tugas 4: Lakukan penggabungan DataFrame dan visualisasikan data dengan Pandas.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Membuat DataFrame Pandas
data_pandas_usia = {"Nama": ["Maya", "Nanda", "Oscar", "Putri", "Qori", "Rudi", "Sinta", "Toni", "Vira", "Wayan", "Yanti", "Zain"],
                    "Usia": [44, 28, 33, 36, 21, 49, 37, 26, 32, 40, 24, 51] }
df_pandas_usia = pd.DataFrame(data_pandas_usia)

# Membuat DataFrame kedua
data_pandas_pekerjaan = {"Nama": ["Maya", "Nanda", "Oscar", "Putri", "Qori", "Rudi", "Sinta", "Toni", "Vira", "Wayan", "Yanti", "Zain"],
                         "Pekerjaan": ["Nelayan", "Akuntan publik", "Koki", "Guru", "Desainer Grafis", "Insinyur", "Buruh", "Penjahit", "Guru", "Pelukis", "Akuntan publik", "Buruh"],}
df_pandas_pekerjaan = pd.DataFrame(data_pandas_pekerjaan)

# Membuat DataFrame Ketiga - Data Hobi disesuaikan dengan nama-nama yang ada
data_pandas_hobi = {"Nama": ["Maya", "Nanda", "Oscar", "Putri", "Qori", "Rudi", "Sinta", "Toni", "Vira", "Wayan", "Yanti", "Zain"],
                     "Hobi": ["Memancing", "Joging", "Memasak", "Menonton Drakor", "Membaca Buku", "Mendaki Gunung", "Menonton Drakor", "Bersepeda", "Bersepeda", "Fotografi", "Joging", "Memancing"]}
df_pandas_hobi = pd.DataFrame(data_pandas_hobi)

# Membuat DataFrame Keempat - Data Gender disesuaikan dengan nama-nama yang ada
data_pandas_gender = {"Nama": ["Maya", "Nanda", "Oscar", "Putri", "Qori", "Rudi", "Sinta", "Toni", "Vira", "Wayan", "Yanti", "Zain"],
                      "Gender": ["Perempuan", "Laki-laki", "Laki-laki", "Perempuan", "Perempuan", "Laki-laki", "Perempuan", "Laki-laki", "Perempuan", "Laki-laki", "Perempuan", "Laki-laki"]}
df_pandas_gender = pd.DataFrame(data_pandas_gender)

# Join
df_joined = pd.merge(df_pandas_usia, df_pandas_pekerjaan, on="Nama")
df_joined = pd.merge(df_joined, df_pandas_hobi, on="Nama")
df_joined = pd.merge(df_joined, df_pandas_gender, on="Nama")
print("DataFrame Gabungan:")
display(df_joined)

# Menghitung statistik deskriptif
print("\nStatistik Deskriptif:")
display(df_joined.describe())

# Plotting Data Usia
df_joined.set_index('Nama')['Usia'].plot(kind='bar')
plt.title('Distribution of Usia')
plt.show()

# Plotting Data Pekerjaan
df_joined['Pekerjaan'].value_counts().plot(kind='bar')
plt.title('Distribution of Pekerjaan')
plt.show()

# Plotting Data Hobi
df_joined['Hobi'].value_counts().plot(kind='bar')
plt.title('Distribution of Hobi')
plt.show()

# Plotting Data Gender
df_joined['Gender'].value_counts().plot(kind='bar')
plt.title('Distribution of Gender')
plt.show()

```

Tugas 5: Gunakan metode ini untuk menggabungkan data yang Anda buat di PySpark dengan data dari Pandas, kemudian lakukan analisis sederhana seperti menghitung rata-rata usia.

```python
import pandas as pd

# Mengonversi df_tugas (PySpark DataFrame) ke Pandas DataFrame
df_tugas_pandas = df_tugas.toPandas()
print("Dataframe konversi:")
display(df_tugas_pandas)

# Menggabungkan df_tugas_pandas dengan df_pandas_tugas
df_gabungan_tugas5 = pd.merge(df_tugas_pandas, df_pandas_tugas, how="outer")

print("DataFrame Gabungan (Tugas 5):")
display(df_gabungan_tugas5)

# Menghitung rata-rata usia dari DataFrame gabungan
rata_rata_usia_gabungan_tugas5 = df_gabungan_tugas5['Usia'].mean()

print(f"\nRata-rata usia dari DataFrame gabungan (Tugas 5): {rata_rata_usia_gabungan_tugas5:.2f}")
```

Tugas 6: Gabungkan data dari PySpark dan Pandas, lalu lakukan operasi statistik seperti menghitung nilai maksimum usia.

```python
# Mengonversi df_tugas (PySpark DataFrame) ke Pandas DataFrame
df_tugas_pandas = df_tugas.toPandas()

# Menggabungkan df_tugas_pandas dengan df_pandas_tugas
df_gabungan_tugas6= pd.merge(df_tugas_pandas, df_pandas_tugas,how="outer")

print("DataFrame Gabungan:")
display(df_gabungan_tugas6)

# Menghitung nilai maksimum usia dari DataFrame gabungan
maksimum_usia_gabungan = df_gabungan_tugas6['Usia'].max()

print(f"\nNilai maksimum usia dari DataFrame gabungan: {maksimum_usia_gabungan}")
```
