```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# create a multi-bar chart to compare data across groups
url = 'https://raw.githubusercontent.com/BagusMukhamadMuslim/PraktikumBigData/refs/heads/main/titanic_processed_final.csv'
df_plot = pd.read_csv(url)

# Pclass vs Survived
plt.figure(figsize=(9, 6))
sns.countplot(
    x='Pclass',
    hue='Survived',
    data=df_plot,
    palette='viridis' 
)
plt.title('1. Tingkat Kelangsungan Hidup Berdasarkan Kelas Penumpang', fontsize=15)
plt.xlabel('Kelas Penumpang', fontsize=12)
plt.ylabel('Jumlah Penumpang', fontsize=12)
plt.xticks(ticks=[0, 1, 2], labels=['Kelas 1', 'Kelas 2', 'Kelas 3'])

handles, labels = plt.gca().get_legend_handles_labels()
plt.legend(handles, ['Tidak Selamat (0)', 'Selamat (1)'], title='Status')
plt.grid(axis='y', linestyle='--')
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(
    x='FamilySize',
    hue='Survived',
    data=df_plot,
    palette='magma' 
)

# FamilySize vs Survived
plt.title('2. Tingkat Kelangsungan Hidup Berdasarkan Ukuran Keluarga', fontsize=15)
plt.xlabel('Ukuran Keluarga (SibSp + Parch + 1)', fontsize=12)
plt.ylabel('Jumlah Penumpang', fontsize=12)

unique_sizes = sorted(df_plot['FamilySize'].unique())
plt.xticks(ticks=range(len(unique_sizes)), labels=unique_sizes, rotation=0)

# Kustomisasi Legenda
plt.legend(handles, ['Tidak Selamat (0)', 'Selamat (1)'], title='Status')
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
# Generate scatter plots with Seaborn to explore relationships between variables

# Hubungan Age dan Fare berdasarkan Survived
plt.figure(figsize=(10, 7))
sns.scatterplot(
    x='Age',
    y='Fare',
    hue='Survived',
    data=df_plot,
    alpha=0.6,
    s=70, 
    palette='coolwarm' 
)

# Kustomisasi Plot
plt.title('Hubungan antara Usia dan Tarif Berdasarkan Kelangsungan Hidup', fontsize=14)
plt.xlabel('Usia (Standardized Z-Score)', fontsize=12)
plt.ylabel('Tarif (Log Transformed)', fontsize=12)

# Kustomisasi Legenda
handles, _ = plt.gca().get_legend_handles_labels()
plt.legend(handles, ['Tidak Selamat (0)', 'Selamat (1)'], title='Status')
plt.grid(axis='both', linestyle='--', alpha=0.7)

plt.show()


# Hubungan Age dan Fare berdasarkan Sex_male
plt.figure(figsize=(10, 7))
sns.scatterplot(
    x='Age',
    y='Fare',
    hue='Sex_male',
    size='Survived', # Ukuran titik berdasarkan Status Selamat
    data=df_plot,
    sizes=(30, 150), # Rentang ukuran titik
    alpha=0.6,
    palette='bwr' # Biru/Putih/Merah
)

# Kustomisasi Plot
plt.title('Usia vs. Tarif: Dipisahkan Berdasarkan Jenis Kelamin dan Kelangsungan Hidup', fontsize=14)
plt.xlabel('Usia (Standardized Z-Score)', fontsize=12)
plt.ylabel('Tarif (Log Transformed)', fontsize=12)

# Kustomisasi Legenda
h, l = plt.gca().get_legend_handles_labels()
# Legenda Hue (Gender)
plt.legend(h[0:2], ['Wanita (0)', 'Pria (1)'], title='Jenis Kelamin', loc='upper right', bbox_to_anchor=(1.25, 1))
# Legenda Size (Survived)
plt.legend(h[2:], ['Tidak Selamat', 'Selamat'], title='Status', loc='lower right', bbox_to_anchor=(1.25, 0))

plt.grid(axis='both', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Customize plot layouts and color schemes.

# Membuat Stripplot (plot Age vs Pclass colored Survived)
sns.set_style("whitegrid") 
plt.figure(figsize=(10, 7)) 

# Define Custom Palette: Merah (Tidak Selamat), Hijau (Selamat)
custom_palette = {0: '#E74C3C', 1: '#2ECC71'}

sns.stripplot(
    x='Pclass',
    y='Age',
    hue='Survived',
    data=df_plot,
    jitter=0.2, 
    dodge=True,
    alpha=0.7,
    palette=custom_palette,
    size=5
)

# Kustomisasi Plot
plt.title('Hubungan Usia dan Kelas Penumpang Terhadap Kelangsungan Hidup', fontsize=15)
plt.xlabel('Kelas Penumpang', fontsize=12)
plt.ylabel('Usia (Standardized Z-Score)', fontsize=12)
plt.xticks(ticks=[0, 1, 2], labels=['Kelas 1', 'Kelas 2', 'Kelas 3']) # Mengganti nilai 1, 2, 3

# Kustomisasi Legenda
handles, labels = plt.gca().get_legend_handles_labels()
plt.legend(handles, ['Tidak Selamat (0)', 'Selamat (1)'], title='Status', loc='upper right')

plt.tight_layout()
plt.savefig('Pclass_Age_Stripplot.png')
plt.show()
```
